```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
f = open('tao_te_ching.txt',encoding = "ISO-8859-1") 
raw_TTC = f.read()
FindWordsText_TTC = nltk.word_tokenize(raw_TTC) 
text_TTC = nltk.Text(FindWordsText_TTC) 
```

```python
len(text_TTC)
```

```python
text_TTC.count('world')
```

```python
text_TTC.concordance("world")
```

```python
text_TTC.similar("world")
```

```python
text_TTC.common_contexts(["the", "world"])
```

```python
text_TTC.dispersion_plot(["Tao", "wise", "nature", "people", "True", "great", "world"])
```

```python
frequency_dist = nltk.FreqDist(text_TTC)
frequency_dist.most_common(25)
```
