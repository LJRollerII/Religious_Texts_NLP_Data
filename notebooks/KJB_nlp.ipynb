```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
f = open('king_james_bible.txt',encoding = "ISO-8859-1") 
raw_KJB = f.read()
FindWordsText_KJB = nltk.word_tokenize(raw_KJB) # Seperate text into words and preprocessing
text_KJB = nltk.Text(FindWordsText_KJB) # Spit out finished, searchable version of text
```

```python
len(text_KJB)
```

```python
text_KJB.count('Christ')
```

```python
text_KJB.concordance("Christ")
```

```python
text_KJB.similar("Christ")
```

```python
text_KJB.common_contexts(["Jesus", "Christ"])
```

```python
text_KJB.dispersion_plot(["Christ", "repentance", "love", "peace", "prayer", "mercy", "God"])
```

```python
frequency_dist = nltk.FreqDist(text_KJB)
frequency_dist.most_common(25)
```
