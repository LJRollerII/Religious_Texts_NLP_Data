```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
f = open('the_quran.txt',encoding = "ISO-8859-1")
raw_Quran = f.read()
FindWordsText_Quran = nltk.word_tokenize(raw_Quran)
text_Quran = nltk.Text(FindWordsText_Quran)
```

```python
len(text_Quran)
```

```python
text_Quran.count('Muhammad')
```

```python
text_Quran.concordance("Muhammad")
```

```python
text_Quran.similar("Muhammad")
```

```python
text_Quran.common_contexts(["Muslims", "Mecca"])
```

```python
text_Quran.dispersion_plot(["Muhammad", "Lord", "believe", "life", "praise", "Mecca", "God"])
```

```python
frequency_dist = nltk.FreqDist(text_Quran)
frequency_dist.most_common(25)
```
